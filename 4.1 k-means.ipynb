uczenie maszynowe dzielimy na nadzorowane (klasyfikacja i regresja) oraz nienadzorowane (klastrowanie i analiza skupień)

```python
import pandas as pd

df = pd.read_csv('Mall_Customers.csv')
```

```python
df
```

```python
#do dalszej analizy bierzemy tylko kolumny od Age włącznie
X_train = df.iloc[:,2:]

#nie tworzymy zbioru y, bo tu ich nie mamy; nie dajemy etykiet 
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(df["Age"], df["Annual Income (k$)"], df["Spending Score (1-100)"])
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
```

```python
#klasteryzacja za pomocą K-means i 4 klastrami
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4) #definiujemy sobie model i określamy liczbę klastrów
model = kmeans.fit(X_train) #tworzymy klastry
labels = model.predict(X_train) #nadajemy każdej próbce jakąś klasę
```

```python
%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(df["Age"], df["Annual Income (k$)"], df["Spending Score (1-100)"], c=labels)

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
```

```python
#zrób to samo dla 7 klastrów i wyświetl wykres

kmeans = KMeans(n_clusters=7) #definiujemy sobie model i określamy liczbę klastrów
model = kmeans.fit(X_train) #tworzymy klastry
labels = model.predict(X_train) #nadajemy każdej próbce jakąś klasę

%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(df["Age"], df["Annual Income (k$)"], df["Spending Score (1-100)"], c=labels)

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
```
