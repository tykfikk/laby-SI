```python
import pandas as pd

df = pd.read_csv('StudentsPerformance.csv')
df
```

```python
import matplotlib.pyplot as plt

plt.plot(df['math score'])
```

co jest nie tak z tym wykresem?
nie ma podpisów osi (w ogóle nie wiadomo co się dzieje na tym wykresie) 
typ wykresu jest nie tak; wyniki są od siebie niezależne, więc powinny być kolumnami 
oglądacz wykresu musi rozumieć wykres elo

```python
plt.plot(df['math score'])
plt.xlabel('Sample')
plt.ylabel('Test Score')
```

teraz dorabiamy legendę (czyli przedstawiamy nie tylko math score)

```python
plt.plot(df['math score'], label = 'math', color = 'mediumpurple')
plt.plot(df['reading score'], label = 'reading', color = 'lightcoral')
plt.plot(df['writing score'], label = 'writing', color = 'yellowgreen')
plt.xlabel('Sample')
plt.ylabel('Test Score')
plt.legend()
```

```python
plt.plot(df['math score'], '*', label = 'math', color = 'mediumpurple')
plt.plot(df['reading score'], '^', label = 'reading', color = 'lightpink')
plt.plot(df['writing score'], '.', label = 'writing', color = 'yellowgreen')
plt.xlabel('Sample')
plt.ylabel('Test Score')
plt.legend()
```

polecenie subplot - na jednym rysunku wyświetlamy więcej wykresów

```python
plt.subplot(1,3,1)
#(ile wierszy wykresów, ile kolumn, który wykres definiujemy)
plt.plot(df['math score'], '*', label = 'math', color = 'mediumpurple')
plt.xlabel('Sample')
plt.ylabel('Test Score')
plt.legend()
plt.subplot(1,3,2) 
plt.plot(df['reading score'], '^', label = 'reading', color = 'lightpink')
plt.xlabel('Sample')
plt.legend()
plt.subplot(1,3,3) 
plt.plot(df['writing score'], '.', label = 'writing', color = 'yellowgreen')
plt.xlabel('Sample')
plt.legend()
plt.savefig('plot2.png')
```

```python
plt.subplot(3,1,1)
#(ile wierszy wykresów, ile kolumn, który wykres definiujemy)
plt.plot(df['math score'], '*', label = 'math', color = 'mediumpurple')
plt.ylabel('Test Score')
plt.legend()
plt.subplot(3,1,2) 
plt.plot(df['reading score'], '^', label = 'reading', color = 'lightpink')
plt.ylabel('Test Score')
plt.legend()
plt.subplot(3,1,3) 
plt.plot(df['writing score'], '.', label = 'writing', color = 'yellowgreen')
plt.xlabel('Sample')
plt.ylabel('Test Score')
plt.legend()
plt.savefig('plot3.png')
```
