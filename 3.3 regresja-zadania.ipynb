regresja - prognozowanie wartości; linia prosta najlpiej odwzorowuje dane na układzie współrzędnych (mamy zbiór punktów i szukamy takiej prostej, która będzie najlepiej pokazywała ich układ) 

miarą dobroci tego aalgorytmu są błędy MSE (średni błąd kwadratowy) i MAE (średni błąd absolutny) - średnie odległości punktów od wyznaczonej prostej 
*MSE bardziej reaguje na obecność krajnych punktów (pokazuje jeżeli nie udało się do nich dopasować prostej)

```python
#wczytaj plik kc_house_data.csv

import pandas as pd

df = pd.read_csv('kc_house_data.csv')
```

```python
#wyświetl plik

df
```

```python
#za pomocą polecenia drop usuń co najmniej kolumny "id" oraz "date"

df = df.drop(['id', 'date'], axis = 1)
```

```python
#wyświetl plik i sprawdź, czy kolumny zostały usunięte
df
```

```python
import numpy as np
y = np.asarray(df[["price"]])
X = np.asarray(df[["sqft_living"]])
```

```python
#narysuj wykres punktowy (plot) zależności pomiedzy X i y
import matplotlib.pyplot as plt
plt.plot(X,y,'*', color = 'darkseagreen')
plt.ylabel("price")
plt.xlabel("sqft_living")
plt.rcParams['figure.figsize'] = (10,7)
```

```python
#podziel dana na treningowe i testowe, przyjmij train_size = 0.8
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(mse, mae)
```
